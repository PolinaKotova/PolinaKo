```
import string
```

```
with open("./city smells.txt") as myfile:
     text = myfile.read()
cs = text.lower().translate(str.maketrans('','',string.punctuation)).replace("«","").replace("»","").split()
csw = set(cs)
w_total = len(cs)
n51 = len(csw)
n52 = n51 / w_total
n52 = round(n52, 2)
print(n51)
print(n52)

sum = 0
for i in cs:
    sum = sum + len(i)

n1 = sum / len(cs)
n1 = round(n1, 2)
print(n1)
```

```
r = open('result.txt', 'w')
r.write("Результаты исследования текста:" + '\n' + "1. Средняя длина слова в тексте равна " + str(n1) + '.\n' + "5. Количество уникальных слов равно " + str(n51) + ", пророрция общему количеству слов в тексте равна " + str(n52) + '.\n')
r.close()
```
